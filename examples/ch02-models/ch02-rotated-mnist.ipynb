```python
import torch
import torchvision
from torch.utils.data import DataLoader
from torch.utils.data import Subset
import matplotlib.pyplot as plt
import numpy as np
```

```python
mnist_data = torchvision.datasets.MNIST(".", train=True, download=True)  
mnist_data_rotated = torchvision.datasets.MNIST(".", train=True, download=True,
                                                  transform=torchvision.transforms.Compose([
                                                    torchvision.transforms.RandomRotation((30,75), fill=(0,))
                                                  ]))  
```

```python
fig = plt.figure()  
indices = np.arange(len(mnist_data.targets))
for i in range(5):
    if i in (0,2):
        image_set  = Subset(mnist_data, indices[mnist_data.targets==i])
    else:
        image_set  = Subset(mnist_data_rotated, indices[mnist_data.targets==i])    
    images = enumerate(image_set)
    for j in range(10):   
        plt.subplot(5, 10, i*10+j+1)
        index, (image, target) = next(images)
        plt.imshow(image, cmap='gray', interpolation='none')
        plt.xticks([])
        plt.yticks([])
```
